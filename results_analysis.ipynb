```python
import numpy as np
import pandas as pd
import ast
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv("results.csv")
results
```

```python
training_loss = np.array(ast.literal_eval(results['training_loss'].iloc[82]))
validation_loss = np.array(ast.literal_eval(results['validation_loss'].iloc[82]))

plt.plot(training_loss, label="Training Loss")
plt.plot(validation_loss, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.legend()
plt.show()
```
